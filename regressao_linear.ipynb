# Regressão linear

Vamos fabricar alguns dados que mostrem uma relação linear (aproximadamente reta) entre a velocidade da página e o valor gasto na compra.

```python
%matplotlib inline
import numpy as np
from pylab import *
```

```python
# Geração de dados
tempo_carregamento = np.random.normal(3.0, 1.0, 1000)
valor_compra = 100 - (tempo_carregamento + np.random.normal(0, 0.1, 1000)) * 3

scatter(tempo_carregamento, valor_compra)
```

Pode-se usar o scipy.state.linregress para tornar as coisas mais simples, já que só existem dois recursos

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(tempo_carregamento, valor_compra)
```

```python
r_value ** 2
```

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept

fitLine = predict(tempo_carregamento)

plt.scatter(tempo_carregamento, valor_compra)
plt.plot(tempo_carregamento, fitLine, c='r')
plt.show()
```
